### *Plots*

*Reading and opening the data*

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# reading the data
crDF = pd.read_csv('C:/Users/ivan2/OneDrive/Documents/School/BU/DSC 640 Data Presentation - Visualization/Sample Datasets/ex4-2/crimerates-by-state-2005.csv')
crDF.head(5)
```

```python
crDF.dtypes
```

 Now Iâ€™ll create the Density Plot.

```python
# filter out the 'United States' record
crDF_filtered = crDF[crDF['state'] != 'United States']

# set up the matplotlib figure
plt.figure(figsize=(8, 6))

# create the density plot with seaborn
sns.kdeplot(crDF_filtered['population'], fill=True, color="blue", alpha=0.5)

# plot labels
plt.title('Python - Density Plot: Density Plot of Population')
plt.xlabel('Population')
plt.ylabel('Density')

# plot
plt.show()
```

 Now Iâ€™ll create the scatterplot.

```python
# set up the figure and create the plot
plt.figure(figsize=(10, 6))
plt.scatter(crDF_filtered['murder'], crDF_filtered['aggravated_assault'])

# plot labels
plt.title('Python - Scatter Plot: Plot of Murder vs Aggravated Assault')
plt.xlabel('Murder Rate (per 100,000)')
plt.ylabel('Aggravated Assault Rate (per 100,000)')

# plot
plt.show()
```

Now I'll create the bubble chart.

```python
# first we need to filter the dataset to drop the D.C. outlier record
crDF_no_outlier = crDF_filtered[crDF_filtered['state'] != 'District of Columbia']

# set up the figure and create the plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(crDF_no_outlier['murder'], crDF_no_outlier['robbery'],
                      s=crDF_no_outlier['population'] / 40000,  # adjusting the wize of bubbles
                      alpha=0.6)

# plot labels
plt.title('Python - Bubble Chart: Chart of Murder vs Robbery with Population Size')
plt.xlabel('Murder Rate (per 100,000)')
plt.ylabel('Robbery Rate (per 100,000)')

# plot
plt.show()
```
