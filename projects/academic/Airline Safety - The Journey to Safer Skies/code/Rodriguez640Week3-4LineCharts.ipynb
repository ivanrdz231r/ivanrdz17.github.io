### *Plots*

*Reading and opening the data*

```python
import pandas as pd

# reading the data
wp_hs = pd.read_excel('C:/Users/ivan2/OneDrive/Documents/School/BU/DSC 640 Data Presentation - Visualization/Sample Datasets/ex2-2/world-population.xlsm')
wp_hs.head(5)
```

```python
import matplotlib.pyplot as plt
```

 Now Iâ€™ll create the line chart for world population.

```python
# Group the data by Year and sum of Population
population_by_year = wp_hs.groupby('Year')['Population'].sum()

# Create a bar graph
plt.figure(figsize=(12, 8))
population_by_year.plot(kind='line', color='skyblue')
plt.title('Python - Line Chart: Sum of Population by Year')
plt.xlabel('Year')
plt.ylabel('Population in Billions')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

 Now Iâ€™ll create the Step chart for world population.

```python
# Group the data by Year and sum of Population
population_by_year = wp_hs.groupby('Year')['Population'].sum()

# Create a step chart
plt.figure(figsize=(12, 8))
population_by_year.plot(kind='line', drawstyle='steps-post', color='skyblue')  # Using 'line' kind with 'steps-post' drawstyle for step chart
plt.title('Python - Step Chart: Sum of Population by Year')
plt.xlabel('Year')
plt.ylabel('Population in Billions')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```
